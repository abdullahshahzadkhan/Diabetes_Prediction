```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

# Loading Data

```python
df = pd.read_csv('/kaggle/input/pima-india/diabetes.csv')
df.head()
```

```python
print(df.shape)
df.describe()
```

```python
df.info()
```

```python
df.Outcome.value_counts()
```

```python
df.groupby('Outcome').mean()
```

# Separate X and Y

```python
X = df.drop(columns = 'Outcome', axis=1)
Y = df.Outcome
```

# Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python
print(standardized_data)
```

```python
X = standardized_data
Y = df.Outcome
```

# Train-Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape, Y.shape, Y_train.shape, Y_test.shape)
```

# Train a Model

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

```python
#Testing on Train Data
X_train_predictions = classifier.predict(X_train)
training_accuracy = accuracy_score(X_train_predictions, Y_train)
training_accuracy
```

```python
#Testing on Test Data
X_test_predictions = classifier.predict(X_test)
test_accuracy = accuracy_score(X_test_predictions, Y_test)
test_accuracy
```

## Trying Simple Log Reg

```python
from sklearn.linear_model import LogisticRegression

# instantiate the model (using the default parameters)
logreg = LogisticRegression(random_state=16)
```

```python
logreg.fit(X_train, Y_train)
```

```python
X_train_predictions = logreg.predict(X_train)
train_accuracy = accuracy_score(X_train_predictions, Y_train)
print('train set accuracy:', train_accuracy)

X_test_predictions = logreg.predict(X_test)
test_accuracy = accuracy_score(X_test_predictions, Y_test)
print('test set accuracy:', test_accuracy)
```

# Making a Predictive System

```python
input_data = (4,110,92,0,0,37.6,0.191,30)

# Changing to np array
input_array = np.asarray(input_data)

input_array.shape
```

```python
prediction = classifier.predict(std_data)
print(prediction)

if prediction == 0:
    print('Non Diabetic')
else:
    print('Diabetic')
```
